```python
df_londom2.describe()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_sport = LabelEncoder()
label_encoder_sex = LabelEncoder()
```

```python
df_londom2['Sport'] = label_encoder_sport.fit_transform(df_londom2['Sport'])
df_londom2['Sex'] = label_encoder_sex.fit_transform(df_londom2['Sex'])
```

```python
df_londom2['Sport_txt'] = label_encoder_sport.inverse_transform(df_londom2['Sport'])
df_londom2['Sex_txt'] = label_encoder_sex.inverse_transform(df_londom2['Sex'])
```

```python
df_londom2.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = 'Height', y = 'Weight', hue = 'Sport_txt', data = df_londom2)
plt.title('Peso x Altura dos Atletas por Esporte')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.legend(bbox_to_anchor=(1.05, 1), loc = 'upper left')
plt.show()
```
