```python
# importando o pandas
import pandas as pd
```

```python
# importando pacotes para graficos
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
path = 'C:/Users/noturno/turma2/datasets/'
```

```python
df_mark = pd.read_csv(path + 'dados_marketing.csv')
```

```python
df_mark.head()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_mark['Visitantes'], bins=10, kde=True, color='blue')
plt.title('Distribuição de Visitantes por Dia')
plt.xlabel('Número de Visitantes')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Nome_Dia', y='Vendas', data=df_mark, color='skyblue')
plt.title('Vendas por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Vendas')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Gastos_Mkt', y='Vendas', hue='Promo', data=df_mark, palette='coolwarm', s=100)
plt.title('Gastos de Marketing vs Vendas')
plt.xlabel('Gastos com Marketing')
plt.ylabel('Vendas')
plt.legend(title='Promoção')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df_mark.groupby('Nome_Dia')['Gastos_Mkt'].sum().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm', legend=True)
plt.title('Proporção dos Gastos de Marketing por Dia')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Nome_Dia', y='Visitantes', data=df_mark, color='skyblue')
plt.title('Distribuição de Visitantes por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Número de Visitantes')
plt.show()
```
